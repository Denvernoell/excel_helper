```python
# PrintType = "Table"
PrintType = "Figure"

def InchesToPoints(inches):
    return inches * 72.0



if PrintType == "Table":
    ps.LeftMargin = InchesToPoints(0.5)
    ps.RightMargin = InchesToPoints(0.5)
    ps.TopMargin = InchesToPoints(0.5)
    ps.BottomMargin = InchesToPoints(0.5)
    ps.HeaderMargin = InchesToPoints(0)
    ps.FooterMargin = InchesToPoints(0)
    ps.CenterHorizontally = True
    ps.CenterVertically = False
    
if PrintType == "Figure":
    ps.LeftMargin = InchesToPoints(0)
    ps.RightMargin = InchesToPoints(0)
    ps.TopMargin = InchesToPoints(0)
    ps.BottomMargin = InchesToPoints(0)
    ps.HeaderMargin = InchesToPoints(0)
    ps.FooterMargin = InchesToPoints(0)
    ps.CenterHorizontally = True
    ps.CenterVertically = False
```

```python
# Orientation
# 1 = Portrait
# 2 = Landscape

# PaperSize
# 8.5 x 11 = 1
# 11 x 17 = 3

ps.Orientation = 2
ps.PaperSize = 3
ps.FitToPagesWide = 1
ps.FitToPagesTall = 1

print(ps.Orientation)
print(ps.PaperSize)
print(ps.FitToPagesWide)
print(ps.FitToPagesTall)



```

```python





End Sub

Sub HeadingIncrease()

Dim Range1 As String
Dim HeadingIncrease As Variant

Application.ScreenUpdating = False

Do Until ActiveCell.value = ""

sheetName = ActiveCell.value
Range1 = ActiveCell.Offset(0, 8).value
HeadingIncrease = ActiveCell.Offset(0, 9).value

If HeadingIncrease <> "" Then

Worksheets(sheetName).Select
Range(Range1).Select
Selection.Font.Size = Selection.Font.Size + HeadingIncrease

ActiveCell.Offset(-1, 0).Select
Selection.Font.Size = Selection.Font.Size + HeadingIncrease


Worksheets("Properties").Select


End If


ActiveCell.Offset(1, 0).Select
Loop

Application.ScreenUpdating = True

End Sub



Sub MarginIncrease()

Dim LeftBorderIncrease As Variant
Dim TopBorderIncrease As Variant
Dim sheetName As String



Do Until ActiveCell.value = ""

sheetName = ActiveCell.value
LeftBorderIncrease = ActiveCell.Offset(0, 5).value
TopBorderIncrease = ActiveCell.Offset(0, 6).value
If LeftBorderIncrease + TopBorderIncrease = 0 Then GoTo NextSht:
   
   
PrintType = "t"

Select Case PrintType
Case "t"
    With Worksheets(sheetName).PageSetup
        .LeftMargin = ActiveSheet.PageSetup.LeftMargin + Application.InchesToPoints(LeftBorderIncrease)
        .RightMargin = Application.InchesToPoints(0.5)
        .TopMargin = ActiveSheet.PageSetup.TopMargin + Application.InchesToPoints(TopBorderIncrease)
        .BottomMargin = Application.InchesToPoints(0.5)
        .HeaderMargin = Application.InchesToPoints(0)
        .FooterMargin = Application.InchesToPoints(0)
        .CenterHorizontally = True
        .CenterVertically = False
    End With
    
    
 Case "f"
    With Worksheets(sheetName).PageSetup
        .LeftMargin = Application.InchesToPoints(0)
        .RightMargin = Application.InchesToPoints(0)
        .TopMargin = Application.InchesToPoints(0)
        .BottomMargin = Application.InchesToPoints(0)
        .HeaderMargin = Application.InchesToPoints(0)
        .FooterMargin = Application.InchesToPoints(0)
        .CenterHorizontally = True
        .CenterVertically = False
    End With
End Select

NextSht:

ActiveCell.Offset(1, 0).Select
Loop

End Sub


Sub PrintingSettingsIndent()

Dim IndentDecrease As Variant
Dim sheetName As String

Do Until ActiveCell.value = ""

sheetName = ActiveCell.value
IndentDecrease = ActiveCell.Offset(0, 8).value
IndentRange = ActiveCell.Offset(0, 9).value

If IndentDecrease = 0 Then GoTo NextSht:
   
Worksheets(sheetName).Select
Range(IndentRange).Select
Selection.IndentLevel = Selection.IndentLevel - IndentDecrease
Worksheets("Properties").Select

NextSht:

ActiveCell.Offset(1, 0).Select
Loop

End Sub


Sub RangeFinder()


Dim CurRange As String

CurSheet = ActiveSheet.name
CurRange = Selection.Address


Worksheets("Properties").Select
Range("B2").Select
Do Until ActiveCell.value = ""
If ActiveCell.value = CurSheet Then
ActiveCell.Offset(0, 9).value = CurRange
Worksheets(CurSheet).Select
ActiveSheet.Next.Select
Exit Sub
End If
ActiveCell.Offset(1, 0).Select
Loop

End Sub





Sub Spacer()

Dim SpaceIncrease As Variant
Dim sheetName As String
Dim SpaceRange As String


Do Until ActiveCell.value = ""

sheetName = ActiveCell.value
SpaceRange = ActiveCell.Offset(0, 9).value

If ActiveCell.Offset(0, 7).value > 0 Then
SpaceIncrease = ActiveCell.Offset(0, 7).value
End If

If ActiveCell.Offset(0, 7).value < 0 Then
SpaceDecrease = ActiveCell.Offset(0, 7).value
End If

If ActiveCell.Offset(0, 7).value = 0 Then GoTo NextSht:


Worksheets(sheetName).Select
Range(SpaceRange).Select
ActiveCell.Select

Level = 1
If SpaceIncrease <> "" Then GoTo SpaceIncrease
If SpaceDecrease <> "" Then GoTo SpaceDecrease
    
    
    
SpaceIncrease:
    
Select Case SpaceIncrease

Case 3
ActiveCell.Formula = "   " & ActiveCell.Formula

Case 2
ActiveCell.Formula = "  " & ActiveCell.Formula

Case 1
ActiveCell.Formula = " " & ActiveCell.Formula

Case Else

End Select


Level = Level + 1
If Level = 2 Then
ActiveCell.Offset(1, 0).Select
GoTo SpaceIncrease
End If
Worksheets("Properties").Select
If Level = 3 Then GoTo NextSht


SpaceDecrease:

Select Case SpaceDecrease

Case -3
Selection.IndentLevel = Selection.IndentLevel - 1
ActiveCell.Formula = " " & ActiveCell.Formula

Case -2
Selection.IndentLevel = Selection.IndentLevel - 1
ActiveCell.Formula = "  " & ActiveCell.Formula

Case -1
Selection.IndentLevel = Selection.IndentLevel - 1
ActiveCell.Formula = "   " & ActiveCell.Formula

End Select

Level = Level + 1
If Level = 2 Then
ActiveCell.Offset(1, 0).Select
GoTo SpaceDecrease
End If
Worksheets("Properties").Select
If Level = 3 Then GoTo NextSht



NextSht:

ActiveCell.Offset(1, 0).Select


Loop


End Sub

Sub PageSetChange()
    Application.PrintCommunication = True

    With ActiveSheet.PageSetup
        .Orientation = xlLandscape
        .PaperSize = xlPaperLetter
        .FitToPagesWide = 1
        .FitToPagesTall = 2
        
    End With

End Sub

```
